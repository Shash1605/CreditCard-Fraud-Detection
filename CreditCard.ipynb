Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
# Loding the dataset to a Pandas Dataframe
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
 credit_card_data.tail()
```

```python
# Dataset information
credit_card_data.info()
```

```python
# Checking the number of missing values
credit_card_data.isnull().sum()
```

```python
# Distribution of Legit transaction & fraudulent transaction
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced

0 --> Normal Transaction
1 --> Fraudulent Transaction

```python
# Seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under Sampling

```python
legit_sample = legit.sample(n=492)
```

Concatinating two Dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

Now the dataset is balanced

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify = Y, random_state=2)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation based on Accuracy Score

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

Decision Tree

```python
model_dt = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth = 6, min_samples_leaf = 8)
```

```python
model_dt.fit(X_train, Y_train)
```

```python
y_pred = model_dt.predict(X_test)
y_pred
```

```python
model_dt.score(X_test, Y_test)
```

```python
print(classification_report(Y_test, y_pred, labels = [0,1]))
```

As the dataset is balanced there is no need for Sampling. 

The accuracy score is 92% and is very much similar to the Logistic Regression model which is 93%
